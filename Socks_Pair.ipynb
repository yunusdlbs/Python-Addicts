```python
socks=int(input("Enter the total number of socks: "))
array=list(input("Enter your color type of socks: ").split())
color=[]
pair=[]
remainder=[]
for i in set(array):
    if array.count(i)==1: color.append(i), remainder.append(array.count(i)), pair.append(0)
    else: 
        remainder.append([array.count(i)%2])
        color.append(i)
        pair.append(int(array.count(i)/2))
        
for i,j in zip(pair,color):
    print("You have {} pairs {} socks.".format(i,j), end=" ")
print("\n")
for k,l in zip(remainder,color):
    print("You also have {} {} socks remainder.".format(k,l),end=" ")
    
```
