**Given an integer n, return true if it is a power of two. Otherwise, return false.**
**An integer n is a power of two, if there exists an integer x such that n == 2x.**






image.png

```python
num=int(input("Enter your number: "))
power_num=num
counter =1
if num==1: print ("True, 1 is the power of 2. 1=2**0")
else:
    while power_num%2==0:
        power_num//=2
        if power_num==1: 
            print("True, your number is the power of 2. {} = 2**{}.".format(num,counter))
            break
        counter+=1
    else:
        print("False, your number isn't the power of 2.")
    
```

```python
n = int(input(""))
n_list = []
for i in range(2,n+1) :
  if n % i == 0 :
    n_list += [i]
all([i % 2 == 0 for i in n_list])
```
